```python
import pandas as pd
```

```python
# Read data from file
df = pd.read_csv('adult.data.csv')
df.head()
```

```python
# How many of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.

race_count = pd.Series(df['race'].value_counts())
race_count
```

```python
# What is the average age of men?
#Can't do average on Series, so sum divide by count
average_age_men = df.loc[df['sex'] == 'Male', 'age'].mean().round(1)
average_age_men
```

```python
df.tail()
```

```python
# What is the percentage of people who have a Bachelor's degree?
percentage_bachelors = round(df.loc[df['education'] == 'Bachelors', 'education'].count() / df['education'].count() * 100 , 1)
percentage_bachelors
```

```python
education = df['education'].value_counts()
education
```

```python
# What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?
# with `Bachelors`, `Masters`, or `Doctorate`
higher_education = df.loc[df['education'].isin(['Masters','Bachelors','Doctorate']), ['education','salary']]
higher_education
```

```python
# percentage with salary >50K
higher_education_rich = round(higher_education.loc[higher_education['salary'] == '>50K', 'salary'].count() / higher_education['salary'].count() * 100, 1)
higher_education_rich
```

```python
# What percentage of people without advanced education make more than 50K?
# without `Bachelors`, `Masters`, or `Doctorate`
lower_education = df.loc[~df['education'].isin(['Masters','Bachelors','Doctorate']), ['education','salary']]
lower_education
```

```python
# percentage with salary >50K
lower_education_rich = round(lower_education.loc[lower_education['salary'] == '>50K', 'salary'].count() / lower_education['salary'].count() * 100, 1)
lower_education_rich
```

```python
#Checking total quantity if match
total_ed = lower_education.count() + higher_education.count() 
total_ed
```

```python
# What is the minimum number of hours a person works per week (hours-per-week feature)?
min_work_hours = df['hours-per-week'].min()
min_work_hours
```

```python
# What percentage of the people who work the minimum number of hours per week have a salary of >50K?
num_min_workers = df.loc[df['hours-per-week'] == min_work_hours, ['hours-per-week','salary']]
num_min_workers.count()
```

```python
#Percentage
rich_percentage = num_min_workers.loc[num_min_workers['salary'] == '>50K', 'salary'].count() / num_min_workers['salary'].count() * 100
rich_percentage
```

```python
# What country has the highest percentage of people that earn >50K?

#Create dict of percentage of high earning for each country 
highest_earning_lists = round(
    df.loc[df['salary'] == '>50K', 'native-country'].value_counts() 
    / 
    df['native-country'].value_counts() * 100, 1).to_dict()
highest_earning_lists
```

```python
#Query country with highest percentage
highest_earning_country = max(highest_earning_lists, key=highest_earning_lists.get)
highest_earning_country
```

```python
#Query the highest percentage
highest_earning_country_percentage = max(highest_earning_lists.values())
highest_earning_country_percentage
```

```python

# Identify the most popular occupation for those who earn >50K in India.

#Create lists of occupation in India for salary >50K
high_occ_lists = df.loc[(df['native-country'] == 'India') & (df['salary'] == '>50K') , ['native-country','occupation','salary']]

#inserting count of value into the table
high_occ_lists_count = df.loc[(df['native-country'] == 'India') & (df['salary'] == '>50K'),'occupation'].value_counts()
high_occ_lists['count'] = high_occ_lists['occupation'].map(high_occ_lists_count)

#remove duplicated row
high_occ_lists = high_occ_lists.drop_duplicates(subset=['occupation'])
high_occ_lists
```

```python
#Simpler way 

#Create index of unique value of occupation in India 
occupation_index = df.loc[(df['native-country'] == 'India') & (df['salary'] == '>50K'),'occupation'].drop_duplicates()

#Create dict of quantity per occupation in India
high_occupation_list = df.loc[(df['native-country'] == 'India') & (df['salary'] == '>50K'),'occupation'].value_counts().to_dict()

#Set both index, and quantity to Pandas Series
occupation_list = pd.Series(index=occupation_index, data=high_occupation_list)
occupation_list
```

```python
# Identify the most popular occupation for those who earn >50K in India.
top_IN_occupation = occupation_list.idxmax()
top_IN_occupation
```
