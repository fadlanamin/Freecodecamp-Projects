```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Import Data
df = pd.read_csv('medical_examination.csv')
df.head()
```

```python
df['overweight'] = df['weight'] / ((df['height'] /100) ** 2 )
df.head(10)
```

```python
# Add 'overweight' column
df['overweight'] = np.where( df['weight'] / ((df['height'] /100) ** 2 ) > 25, 1, 0)
df.head(10)
```

```python
# Normalize data by making 0 always good and 1 always bad. If the value of 'cholesterol' or 'gluc' is 1, make the value 0. 
# If the value is more than 1, make the value 1.

#Cholesterol & GLuc
df['cholesterol'] = np.where( df['cholesterol'] == 1, 0, 1)
df['gluc'] = np.where( df['gluc'] == 1, 0, 1)

#Using mask func
df.head(10)
```

```python
# Draw Categorical Plot

# Create DataFrame for cat plot using `pd.melt` using just the values from 'cholesterol', 'gluc', 'smoke', 'alco', 'active', and 'overweight'.
#Authors note : melts basically ngegabungin per column nya, jadi kebawah
df_cat = pd.melt(df, value_vars=['cholesterol','gluc', 'smoke', 'alco', 'active', 'overweight'])
df_cat
```

```python
# Group and reformat the data to split it by 'cardio'. 
# You will have to rename one of the columns for the catplot to work correctly.
df_cat = pd.melt(df, id_vars=['cardio'], value_vars=['cholesterol','gluc', 'smoke', 'alco', 'active', 'overweight'])
df_cat
```

```python
#Show the counts of each feature. 
#Authors note :  .reset_index() -- buat ngerubah hasil value count jadi dataframe. 
df_cat = pd.melt(df, id_vars=['cardio'], value_vars=['cholesterol','gluc', 'smoke', 'alco', 'active', 'overweight']).value_counts().reset_index()
df_cat
```

```python
# You will have to rename one of the columns for the catplot to work correctly.
df_cat.rename(columns={df_cat.columns[3]: 'total'}, inplace=True)

#sorting the data by alphabetical
df_cat.sort_values(by='variable', inplace=True)
df_cat
```

```python
# Draw the catplot with 'sns.catplot()'
#y = total, x = variable but split
cat_plot = sns.catplot( data=df_cat, x='variable', y='total', kind='bar', col='cardio' , hue='value' , palette='Set1' )
cat_plot.set_axis_labels('variable', 'total')
cat_plot
```

```python
def draw_heat_map1():
    # Clean the data
    df_heat = pd.DataFrame(df)
    drop_data = df_heat[ (df_heat['weight'] >= df_heat['weight'].quantile(0.975))  #5 weight is more than the 97.5th percentile
               | (df_heat['weight'] <= df_heat['weight'].quantile(0.025)) #4 weight is less than the 2.5th percentile
               | (df_heat['height'] >= df_heat['height'].quantile(0.975)) #3 height is more than the 97.5th percentile
               | (df_heat['height'] <= df_heat['height'].quantile(0.025)) #2 height is less than the 2.5th percentile
               | (df_heat['ap_lo'] >= df_heat['ap_hi'])].index #1 diastolic pressure is higher than systolic ap_lo > ap_hi

    df_heat = df_heat.drop(drop_data)

    # Calculate the correlation matrix
    corr = df_heat.corr().round(1)

    # Generate a mask for the upper triangle
    mask = np.triu(np.ones_like(corr))

    # Set up the matplotlib figure
    fig, ax = plt.subplots()

    # Draw the heatmap with 'sns.heatmap()'
    sns.heatmap(corr, annot=True, square=True, center=0, annot_kws={'fontsize':7 }, linewidths=0.5, mask=mask)

    # Do not modify the next two lines
    fig.savefig('heatmap.png')
    return fig
```

```python
draw_heat_map1()
```
